```python
import acquire_codeup
import prepare_codeup

import unicodedata
import re
import json

import nltk
from nltk.tokenize.toktok import ToktokTokenizer
from nltk.corpus import stopwords

import pandas as pd
```

```python
codeup_df = acquire_codeup.get_blog_articles()
codeup_df.head()
```

```python
news_df = acquire_codeup.get_inshorts_articles()
news_df.head()
```

```python
clean_content = pd.DataFrame([prepare_codeup.basic_clean(i) for i in codeup_df.content])
clean_content.head()
```

```python
codeup_df['clean'] = clean_content
codeup_df.head()
```

```python
stem_clean = pd.DataFrame([prepare_codeup.stem(i) for i in codeup_df.clean])
stem_clean.head()
```

```python
codeup_df['stemmed'] = stem_clean
codeup_df.head()
```

```python
lemmy_clean = pd.DataFrame([prepare_codeup.lemmatize(i) for i in codeup_df.clean])
lemmy_clean.head()
```

```python
codeup_df['lemmatized'] = lemmy_clean
codeup_df.head()
```
